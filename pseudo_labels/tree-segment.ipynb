```python
def annotate_img(repo_path : str, img_name : str, output_dir : Path, confidence : int = 40):
  prediction = model.predict(repo_path+img_name, confidence=confidence).json()
  print("Predicting image ", img_name, ".......")
  image_width = float(prediction['image']['width'])
  image_height = float(prediction['image']['height'])
  predictions = prediction['predictions']
  annotations = []

  for pred in predictions:
    if pred['class'] == "Tree":
      points = pred['points']
      normalized_points = [(point['x'] / image_width, point['y'] / image_height) for point in points]
      points_str = ' '.join([f"{x:.6f} {y:.6f}" for x, y in normalized_points])
      annotation = f"{pred["class_id"]} {points_str}"
      annotations.append(annotation)

  output_path = output_dir / f"{img_name.split('.')[0]}.txt"
  with open(output_path, 'w') as f:
    for annotation in annotations:
      f.write(annotation + '\n')
```

```python
repo_path = "images/"
! mkdir "labels"
output_dir = Path("labels/")

for i, file_path in enumerate(Path(repo_path).glob("*")):
  if file_path.suffix.lower() in [".jpg", ".jpeg",]:
    #src_file = str(file_path)
    #! cp {src_file} {"tmp/"+file_path.name}
    annotate_img(repo_path, file_path.name, output_dir, 25)
print(f"Annotations done for {i} images")
```
