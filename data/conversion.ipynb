```python
'''
Algorithme de transformation du dataset NeonTreeDataset
Conversion des fichiers tiff en png
et mise à jour des noms de fichiers dans les annotations
'''
```

```python
from PIL import Image
import os

# Répertoires
input_dir = "D:/5914554/evaluation/RGB"  # Répertoire des fichiers TIFF
output_dir = "D:/5914554/evaluation/PNG"  # Répertoire de sortie pour les fichiers PNG/JPEG

# Créez le répertoire de sortie s'il n'existe pas
os.makedirs(output_dir, exist_ok=True)

# Conversion des fichiers TIFF en PNG/JPEG
for filename in os.listdir(input_dir):
    if filename.endswith(".tiff") or filename.endswith(".tif"):
        img = Image.open(os.path.join(input_dir, filename))
        output_filename = os.path.splitext(filename)[0] + ".png"  # ou ".jpeg" si vous préférez
        img.save(os.path.join(output_dir, output_filename))
        print(f"Converti {filename} en {output_filename}")

print("Conversion terminée.")
```

```python
import os
import xml.etree.ElementTree as ET

# Répertoires
annotations_dir = "D:/5914554/annotations/annotations"  # Répertoire des fichiers XML
images_dir = "D:/5914554/training/PNG"  # Répertoire contenant les nouvelles images PNG/JPEG

# Extension des nouvelles images (".png" ou ".jpeg")
new_extension = ".png"

# Mise à jour des chemins dans les fichiers XML
for xml_file in os.listdir(annotations_dir):
    if xml_file.endswith(".xml"):
        xml_path = os.path.join(annotations_dir, xml_file)
        
        # Chargement du fichier XML
        tree = ET.parse(xml_path)
        root = tree.getroot()
        
        # Recherche et mise à jour du chemin dans l'élément "filename"
        for filename in root.findall('filename'):
            base_name = os.path.splitext(filename.text)[0]
            new_filename = base_name + new_extension
            filename.text = new_filename
                
        # Sauvegarde du fichier XML avec les chemins mis à jour
        tree.write(xml_path, encoding="utf-8", xml_declaration=True)
        print(f"Mis à jour {xml_file} avec les nouveaux chemins.")

print("Mise à jour des chemins terminée.")
```
